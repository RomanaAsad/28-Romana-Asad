```
from scipy.optimize import linprog
```

```
obj=[4,1]
```

```
lhs_ineq=[[-3,-4],[-1,-5]]
```

```
rhs_ineq=[-20,-15]
```

```
bnd=[(0,float("inf")),(0,float("inf"))]
```

```
from scipy.optimize import linprog
```

```
opt=linprog(c=obj,A_ub=lhs_ineq, b_ub=rhs_ineq,bounds=bnd,method="interior-point")
```
